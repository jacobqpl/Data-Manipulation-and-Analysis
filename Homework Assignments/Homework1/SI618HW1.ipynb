### Q11: Discretize the ratings. (1,2,3,4,5). (15 points)

```python
bins = [0,1,2,3,4,5]

book_data['rating_category'] = pd.cut(
    book_data['book_rating'], bins, 
    labels=['1','2','3','4','5'])

book_data.head()
```

### Q12: For each of the discretized ratings (1,2,3,4,5), what is the average number of ratings? (5 points)

```python
book_data.rating_category.value_counts().mean()
```

### Q13: How about the average number of authors? (5 points)

```python
aver_author = book_data.groupby('rating_category')['book_pages'].mean()
aver_author
```

## Genres: (35 points)

### Q14: Create a new DataFrame, exploding the rows with multiple genres such that it is one row per genre/book. (15 points)

```python
book_data['book_pages']=book_data['genres'].astype('object')
```

```python
bk_data=book_data.drop('genres', axis=1).join(book_data['genres'].str.split('|', expand=True).stack().reset_index(level=1, drop=True).rename('genres'))
```

```python
bk_data
```

### Q15: What is the average rating of books from different genres? What is the median? (10 points)

```python
aver_rating = bk_data.groupby('genres')['book_rating'].mean()
aver_rating
```

```python
bk_data.groupby('genres')['book_rating'].median()
```
